Quick practice making a custom OHE

```python
df = pd.DataFrame({'x':[1,2,3], 'z':['b','c','d']})
df
```

```python
def custom_ohe(df, cols):
    '''
    Input a dataframe and a list of the columns to OHE.
    Return new dataframe with ohe columns.
    '''
    new_df = df.copy()
    for col in cols:
        unique_vals = np.unique(df[col])
        for unique in unique_vals:
            new_df[unique] = (new_df[col] == unique).astype(int)
        new_df.drop(col, axis=1, inplace=True)
    return new_df
```

```python
ohe_df = custom_ohe(df, ['z'])
ohe_df
```
